# Plot interactive maps with the Python folium package

Based on the example from https://python-graph-gallery.com/312-add-markers-on-folium-map/. Interactive maps are especially useful for exploring your data or figuring out which data point ended up in the middle of the ocean.

## Load dependencies

If it says package not available, look for it at anaconda.org and install

```python
import folium
import pandas as pd
```

## Import lat long data into pandas dataframe

```python
dflatlong = pd.read_csv("example_localitydata.csv", header=0) #header = 0 means the first row is a header row, python starts counting at 0
```

## Create and display a folium map centered on OÊ»ahu

```python
m = folium.Map(location=[21.34470, -157.8473], tiles="OpenStreetMap", zoom_start=8)
m
```

## Add locality markers from lat long data in dataframe

The example uses the information from the 'loc_code' column to show when a marker is clicked, you can adjust this as you wish.

```python
for i in range(0,len(dflatlong)):
   folium.Marker(
      location=[dflatlong.iloc[i]['decimallatitude'], dflatlong.iloc[i]['decimallongitude']],
      popup=dflatlong.iloc[i]['loc_code'],
   ).add_to(m)

# Show the map again
m
```

## Save as a standalone html

```python
m.save('interactivemap.html')
```
