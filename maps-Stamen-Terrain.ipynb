# Plot maps with lat long markers and Stamen-terrain background

For the most part this script is very similar to the ESRI shaded relief one but it requires an API key to download the maps. Stamen maps (https://maps.stamen.com/#terrain) was taken over by Stadia in 2023 (https://maps.stamen.com/stadia-partnership/) and Stadia now hosts the Stamen map tiles. Use of the maps for non-commercial purposes is still free, but it requires a free Stadia account and an API key. This will allow you to make 200,000 maps/month. See FAQ about terms of use here: https://stadiamaps.com/faqs/#acceptable-use - they require you to use the appropriate attribution when publishing these maps.

You can create a Stadia account here: https://stadiamaps.com/stamen/onboarding/

The process to set up a "property" (virtual project) and API key for that property is explained here: https://docs.stadiamaps.com/authentication/#api-keys

## Load dependencies

```python
import matplotlib
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cartopy
import pandas as pd
import numpy as np
import cartopy.io.img_tiles as cimgt
from pathlib import Path

# this next magic line plots the maps in the notebook
%matplotlib inline
```

## Create a class to download Stamen terrain background image

Replace API_KEY below with your API key.

The terrain-background is probably the map you will use the most; it is the stamen-terrain map but without city markers or roads. But check out the Stamen watercolor too (https://maps.stamen.com/#watercolor) ;)

```python
class StadiaStamen(cimgt.Stamen):
    def _image_url(self, tile):
        x, y, z = tile
        url = f"https://tiles.stadiamaps.com/tiles/stamen_terrain_background/{z}/{x}/{y}.png?api_key={'API_KEY'}"
        return url
```

## Function to plot map of main HI islands, plot markers, and save to file

```python
def esrimap(dflatlong):
    data_crs = ccrs.PlateCarree() # data to plot is in lat/lon coordinate system
    fig = plt.figure(figsize=(5, 5), dpi= 300, edgecolor='white')
    ax = plt.axes(projection=ShadedReliefESRI().crs, frame_on=False)
    # get bounding box coords here: http://boundingbox.klokantech.com, select 'dublincore'
    ax.set_extent([-160.75, -154.64, 18.78, 22.4]) # minx (westlimit), maxx (eastlimit), miny (southlimit), maxy (northlimit)
    stamen_terrain = StadiaStamen("terrain-background")
    ax.add_image(stamen_terrain, 8) # zoom level 8, you can play with this
    
    # plot points
    listlat = list(dflatlong['decimallatitude'])
    listlong = list(dflatlong['decimallongitude'])
    if listlat: # returns true if the latitude list is not empty and runs the next line
        ax.plot(listlong, listlat, transform=data_crs,
                marker='^',
                color='#f1a340',
                markersize=3,
                zorder=27,
                linestyle='None',
                markeredgewidth=.2,
                markeredgecolor='black',
                alpha=.8)

    # Save to file. If the next lines are active it won't plot the map in the notebook)
    #plt.savefig("mymap.png", dpi=600, bbox_inches="tight")
    #plt.savefig("mymap.pdf", format="pdf", bbox_inches="tight")
    #plt.close('all')
```

## Import csv file with lat long data into pandas dataframe and run mapping function

```python
dflatlong = pd.read_csv("example_localitydata.csv", header=0) #header = 0 means the first row is a header row, python starts counting at 0

try:
    esrimap(dflatlong)
except Exception as e:
    print(e)
    pass
```
